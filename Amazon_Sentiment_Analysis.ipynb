```python
import pandas as pd

# Load the dataset (replace 'dataset.csv' with the actual filename)
df = pd.read_csv("amazon.csv")

# Display first few rows
print(df.head())
```

```python
print(df['review_content'].isnull().sum())
```

```python
df['review_content'].fillna("", inplace=True)
```

```python
from textblob import TextBlob

def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity < 0:
        return "Negative"
    else:
        return "Neutral"

# Apply sentiment function to the reviews
df["sentiment"] = df["review_content"].apply(get_sentiment)

# Show results
print(df[["review_content", "sentiment"]].head())
```

```python
import matplotlib.pyplot as plt

# Count the number of sentiments
sentiment_counts = df["sentiment"].value_counts()

# Plot a bar chart
plt.figure(figsize=(6, 4))
sentiment_counts.plot(kind="bar", color=["green", "red", "blue"])
plt.title("Sentiment Distribution of Product Reviews")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
df.to_csv("sentiment_analysis_results.csv", index=False)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("amazon.csv")

# Check for missing values
df["review_content"].fillna("", inplace=True)

# Display first few rows
print(df[["review_content"]].head())
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Download VADER if not available
nltk.download("vader_lexicon")

# Initialize VADER
sia = SentimentIntensityAnalyzer()

# Function to get sentiment
def get_vader_sentiment(text):
    score = sia.polarity_scores(text)["compound"]
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

# Apply sentiment function
df["vader_sentiment"] = df["review_content"].apply(get_vader_sentiment)

# Display results
print(df[["review_content", "vader_sentiment"]].head())
```

```python
from transformers import pipeline

# Load the sentiment analysis pipeline
sentiment_pipeline = pipeline("sentiment-analysis")

# Function to process in batches
def get_bert_sentiment(text):
    try:
        result = sentiment_pipeline(text[:512])  # Limit to 512 characters
        return result[0]["label"]
    except Exception as e:
        return "Error"  # Handle errors gracefully

# Apply function in batches
df["bert_sentiment"] = df["review_content"].apply(get_bert_sentiment)

# Display results
print(df[["review_content", "bert_sentiment"]].head())
```

```python
import matplotlib.pyplot as plt

# Count sentiment occurrences
vader_counts = df["vader_sentiment"].value_counts()
bert_counts = df["bert_sentiment"].value_counts()

# Plot comparison
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# VADER Sentiment
vader_counts.plot(kind="bar", color=["green", "red", "blue"], ax=axes[0])
axes[0].set_title("VADER Sentiment Analysis")

# BERT Sentiment
bert_counts.plot(kind="bar", color=["green", "red", "blue"], ax=axes[1])
axes[1].set_title("BERT Sentiment Analysis")

plt.show()
```

```python
df.to_csv("advanced_sentiment_results.csv", index=False)
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("advanced_sentiment_results.csv")

# Count sentiment per category
category_sentiment = df.groupby(["category", "vader_sentiment"]).size().unstack().fillna(0)

# Normalize values to percentages
category_sentiment_percentage = category_sentiment.div(category_sentiment.sum(axis=1), axis=0) * 100

# Display sentiment trends per category
print(category_sentiment_percentage)
```

```python
import matplotlib.pyplot as plt

# Plot sentiment distribution per category
category_sentiment_percentage.plot(kind="bar", stacked=True, figsize=(12, 6), colormap="coolwarm")

plt.title("Sentiment Distribution by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Percentage of Reviews")
plt.legend(title="Sentiment")
plt.xticks(rotation=45)
plt.show()
```

```python
# Identify best categories (highest % of positive reviews)
best_categories = category_sentiment_percentage["Positive"].sort_values(ascending=False).head(5)

# Identify worst categories (highest % of negative reviews)
worst_categories = category_sentiment_percentage["Negative"].sort_values(ascending=False).head(5)

print("Top 5 Best Categories (High Positive Sentiment):")
print(best_categories)

print("\nTop 5 Worst Categories (High Negative Sentiment):")
print(worst_categories)
```
